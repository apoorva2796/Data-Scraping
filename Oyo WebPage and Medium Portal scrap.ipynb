## OYO Rooms

```python
url = 'https://www.oyorooms.com/oyos-in-bangalore'
browser.get(url)
```

```python
def get_attribute(parent,selector,attribute):
    try:
        if attribute == 'text':
            val = parent.find_element_by_css_selector(selector).text
        else:
            val = parent.find_element_by_css_selector(selector).get_attribute(attribute)
    except:
        val = None
    return val
```

```python
df_room = pd.DataFrame()
#parent = browser.find_element_by_css_selector('div.hotelgridcontainer__resultsWrapper__inner')
child = browser.find_elements_by_css_selector('div.oyo-row.oyo-row--no-spacing.hotelCardListing')

for room in child:
    name = get_attribute(room,'h3.oyo-cell--12-col.d-secondaryHeading.d-lh24.d-textEllipsis','text')
    location = get_attribute(room,'div.oyo-cell--12-col.d-text14-dark.d-lh20.d-textEllipsis','text')
    price_final = get_attribute(room,'span.listingPrice__finalPrice','text')
    rating = get_attribute(room,'span.hotelRating__rating','text')
    rating = str(rating).split(' ')[0]
    no_of_rating = get_attribute(room,'span.hotelRating__ratingSummary.hotelRating__rating--clickable','text')
    image_url = get_attribute(room,'img.listingImageCard__carouselImg','src')
    room_dict = {'Name':name,
                 'Location':location,
                 'Price Final':price_final,
                 'Rating':rating,
                 'No Of Rating':no_of_rating,
                 'Image URl':image_url
                }
    df_room = df_room.append(room_dict,ignore_index = True)
    
df_room
```

## Medium Portal

```python
import urllib.request as ur
import os
import glob
```

```python
url = 'https://medium.com/search?q=data%20science'
browser.get(url)
```

```python
article_df = pd.DataFrame()
article_parent = browser.find_element_by_css_selector('div.js-postListHandle')
articles = article_parent.find_elements_by_css_selector('div.u-paddingTop20')
for article in articles:
    try:
        article_author = article.find_element_by_css_selector('a.ds-link.ds-link--styleSubtle').text
        article_name = article.find_element_by_css_selector('h3.graf.graf--h3').text
        article_desc = article.find_element_by_css_selector('a.button.button--smaller').get_attribute('href')
        date_posed = article.find_element_by_css_selector('time').text
    except:
        article_author = None
        article_name = None
        article_desc = None
        date_posed = None
        
    article_dict = {
        'Author' : article_author,
        'Name' : article_name,
        'Desc href' : article_desc,
        'Date' : date_posed
    }
    article_df = article_df.append(article_dict,ignore_index = True)
article_df
```

```python
for i in range(len(article_df)):
    url = article_df.loc[i,'Desc href']
    browser.get(url)
    parent = browser.find_element_by_css_selector('div.postArticle-content.js-postField.js-notesSource')
    ptags_text = ''
    each_article = parent.find_elements_by_css_selector('section.section.section--body')
    for each in each_article:
        ptag = each.find_elements_by_css_selector('p')
        for tag in ptag:   ## Extracting paragraphy data
            text = tag.text
            ptags_text = ptags_text + ' ' + text
            
    filename = article_df.loc[i,'Author'] + '(' + str(i) +').txt'

    file = open(filename,"w", encoding='utf-8')
    file.write(str('b'+ptags_text))
```
