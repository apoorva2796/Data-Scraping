```python
from selenium import webdriver
browser = webdriver.Chrome('chromedriver.exe')
```

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://www.goodreads.com/choiceawards/best-books-2017'
browser.get(url)
```

```python
book_df = pd.DataFrame()
book_category = browser.find_element_by_css_selector('div.categoryContainer')
books = book_category.find_elements_by_css_selector('div.category.clearFix')
for book in books:
    name_category = book.find_element_by_css_selector('h4.category__copy').text
    name_book = book.find_element_by_css_selector('img.category__winnerImage').get_attribute('alt')
    URL_image = book.find_element_by_css_selector('img.category__winnerImage').get_attribute('src')
    URL_desc = book.find_element_by_css_selector('a').get_attribute('href')
    book_dict = {
        'Category':name_category,
        'Book_Name':name_book,
        'Image_URL':URL_image,
        'Book_Desc':URL_desc
    }
    book_df = book_df.append(book_dict,ignore_index = True)
ls = ['Category','Book_Name','Image_URL','Book_Desc']
book_df[ls].to_csv('Books.csv',index = False)
book_df
```

```python
books_df = pd.DataFrame()
for i in range(len(book_df)):
    url = book_df.loc[i,'Book_Desc']
    browser.get(url)
    
    parent = browser.find_element_by_css_selector('div.pollContents')
    book = parent.find_element_by_css_selector('div.winningBook')
    Author_name = book.find_element_by_css_selector('span[itemprop="name"]').text
    Short_desc = book.find_element_by_css_selector('div#description.readable').text
    vote_winner = book.find_element_by_css_selector('span.greyText.gcaNumVotes').text
    total_votes = parent.find_element_by_css_selector('div.greyText.gcaNumVotes').text
    vote = int((vote_winner.split(" ")[0]).replace(',',''))
    total = int((total_votes.split(" ")[0]).replace(',',''))
    percent = round(((vote/total)*100),2)
    dict_book = {'Author_name':Author_name,'Short_desc':Short_desc,'vote_winner':vote_winner,'total_votes':total_votes,
                'Percentage':percent,'url':url}
    books_df = books_df.append(dict_book,ignore_index = True)
ls = ['Author_name','Short_desc','vote_winner','total_votes','Percentage']
#books_df[ls].to_csv('Each_book.csv',index = False)
books_df
```

```python
final_df = pd.merge(book_df,books_df,left_on='Book_Desc',right_on='url')
ls = ['Category','Book_Name','Image_URL','Book_Desc','Author_name','Short_desc','vote_winner','total_votes','Percentage']
final_df = final_df[ls]
#final_df.to_csv('Final_books.csv',index = False)
final_df[ls]
```
