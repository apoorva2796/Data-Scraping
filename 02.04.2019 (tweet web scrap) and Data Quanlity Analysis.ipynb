```python
salaries = naukri[~naukri['salary'].str.contains('Not disclosed')]
salaries.head()
```

```python
salaries['salary_new'] = salaries['salary'].str.replace(',','')
salaries['salary_new']
```

```python
import re
list1 = list(salaries['salary_new'])
for sal in list1: 
    ls = re.findall('[0-9]+ - [0-9]+ P.A.',sal)
    print(ls)
#salaries['new_salary']
```

```python
salaries['new_sal'] = salaries['salary_new'].apply(lambda x:re.findall(' [0-9]+ - [0-9]+ P.A.',x) 
                                                     if len(re.findall(' [0-9]+ - [0-9]+ P.A.',x)) > 0
                                                     else 'nan')
salaries['new_sal'].head()
```

```python
salaries['min_sal'] = salaries['new_sal'].apply(lambda x: float('nan') if x == 'nan' 
                                                else str(x).split(' ')[1]).astype(float)
salaries['max_sal'] = salaries['new_sal'].apply(lambda x: float('nan') if x == 'nan' 
                                                else str(x).split(' ')[-2]).astype(float)

ls = ['new_sal','min_sal','max_sal']
salaries[ls].head()
```

```python
## Alternative

salaries['is_salary'] = salaries['salary_new'].str.contains('\d+ - \d+ P.A.')
salaries['min'] = salaries.apply(lambda row: row.salary_new.split(' - ')[0]
                                if row.is_salary else 'nan', axis = 1).astype(float)
salaries['max'] = salaries.apply(lambda row: row.salary_new.split(' - ')[1].split(' P.A.')[0]
                                if row.is_salary else 'nan', axis = 1).astype(float)
```

```python
def exp_range(exp):
    if exp == 0:
        return 'fresher'
    elif exp > 0 & exp <=2:
        return '0-2'
    elif exp > 2 & exp <=5:
        return '2-5'
    elif exp > 5 & exp <=10:
        return '5-10'
    elif exp > 10 & exp <=20:
        return '10-20'
    else:
        return 'Above 20'
salaries['exp_range'] = salaries['min_exp'].apply(exp_range)
salaries.groupby(['exp_range'])['min'].mean()/100000
```

```python
bins = [0.1,2,5,10,20]
# bin 0 : 0
# bin 1 : 0.1 >= x < 2
# bin 2 : 2 >= x < 5
# bin 3 : 5 >= x < 10
# bin 4 : 10 >= x < 20
# bin 5 : x > 20

salaries['exp_range'] = salaries['min_exp'].apply( lambda x: np.digitize(x, bins = bins)).replace({
    0: 'Freshers',
    1: '0-2',
    2: '2-5',
    3: '5-10',
    4: '10-20',
    5: 'Above 20'
})
((salaries.groupby(['exp_range'])['min','max'].mean()/100000).sort_values(by='min').reset_index()).set_index('exp_range')
```
