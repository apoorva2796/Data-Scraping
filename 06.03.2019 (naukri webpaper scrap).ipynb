```python
from selenium import webdriver
browser = webdriver.Chrome('chromedriver.exe')
```

```python
browser.get('http://localhost:8000/book.html')
```

```python
import pandas as pd
import numpy as np
```

```python
df_paras = pd.DataFrame()
books_tag = browser.find_element_by_css_selector('div.books')
book_tag = books_tag.find_elements_by_css_selector('div.book')
for book in book_tag:
    curr_book = {'Author' : book.find_element_by_css_selector('div.author').text,
                 'Desc' : book.find_element_by_css_selector('div.desc').text,
                 'Price' : book.find_element_by_css_selector('li[type = "price"]').text,
                 'Title' : book.find_element_by_css_selector('li.title').text,
                 'Category' : book.find_element_by_css_selector('div.category').get_attribute('type')
                }
    df_paras = df_paras.append(curr_book,ignore_index = True)
    
df_paras
```

```python
url = 'https://www.naukri.com/data-science-jobs'
browser.get(url)
```

```python
df_job = pd.DataFrame()
parents_jobs_tags = browser.find_elements_by_css_selector('div.row[type="tuple"]')

for job in parents_jobs_tags:
    job_id = job.get_attribute('id')
    job_title = job.find_element_by_css_selector('li.desig').get_attribute('title')
    job_company = job.find_element_by_css_selector('span.org').text
    job_experience = job.find_element_by_css_selector('span.exp').text
    job_location = job.find_element_by_css_selector('span.loc').text
    job_keyskill = job.find_element_by_css_selector('span.skill').text
    job_salary = job.find_element_by_css_selector('span.salary').text
    job_date = job.find_element_by_css_selector('span.date').text
    job_url = job.get_attribute('data-url')
    
    job_dict = {'Job URL':job_url,'Job Id':job_id,'Title':job_title,'Company':job_company,'Experience':job_experience,
         'Location':job_location,'Key Skill':job_keyskill,'Salary':job_salary,'Date':job_date}
    df_job = df_job.append(job_dict,ignore_index = True)
    
df_job
```

```python
df_job.to_csv('Jobs_datascience.execl',index = False)
```

# Looping through multiple page

```python
def get_attribute(parent,selector,attribute):
    try:
        if attribute == 'text':
            val = parent.find_element_by_css_selector(selector).text
        else:
            val = parent.find_element_by_css_selector(selector).get_attribute(attribute)
    except:
        val = None
    return val

import time
jobs_dt = pd.DataFrame()
for i in range(1,10):
    url = 'https://www.naukri.com/data-science-jobs-%d' %(i)
    browser.get(url)
    time.sleep(2)
    print(i)
    
    parents_jobs_tags = browser.find_elements_by_css_selector('div.row[type="tuple"]')
   
    for job in parents_jobs_tags:
        job_dt = {
             'Job URL':job.get_attribute('data-url'),
             'Job Id':job.get_attribute('id'),
             'Title':get_attribute(job,'li.desig','title'),
             'Company':get_attribute(job,'span.org','text'),
             'Experience':get_attribute(job,'span.exp','text'),
             'Location':get_attribute(job,'span.loc','text'),
             'Key Skill':get_attribute(job,'span.skill','text'),
             'Salary':get_attribute(job,'span.salary','text'),
             'Date':get_attribute(job,'span.date','text')
         }
        jobs_dt = jobs_dt.append(job_dt,ignore_index = True)
    
jobs_dt
```
